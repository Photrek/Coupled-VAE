Available Methods

Probability density function.   
*pdf(x, loc=None, shape=1, df=1, allow_singular=False)*

Log of the probability density function.   
*logpdf(x, loc=None, shape=1, df=1, allow_singular=False)*

Draw random samples from a multivariate t-distribution.   
*rvs(loc=None, shape=1, df=1, size=1, random_state=None)*   


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy

from scipy.stats import multivariate_t as mvt 
from scipy.stats import multivariate_normal as mvn

x, y = np.mgrid[-1:3:.01, -2:1.5:.01]
pos = np.dstack((x, y))
rv = mvt([1.0, -0.5], [[2.1, 0.3], [0.3, 1.5]], df=15)
fig, ax = plt.subplots(1, 1)
ax.set_aspect('equal')
plt.contourf(x, y, rv.pdf(pos))
plt.title("contourplot scipy.stats.multivariate_t")
```

```python
data = rv.rvs(size=int(1e6))
x, y = data[:, 0], data[:, 1]
plt.scatter(x,y)
plt.title("scatterplot scipy.stats.multivariate_t")
```

```python
plt.hist2d(x,y, bins=100)
plt.title("2d histogram scipy.stats.multivariate_t")
```
